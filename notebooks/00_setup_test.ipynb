```python
import zipfile, os

zip_path = '/content/idrid_raw/indian-diabetic-retinopathy-image-dataset.zip'
extract_path = '/content/idrid_extracted'

os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as z:
    z.extractall(extract_path)

# Yapiyi gor
for root, dirs, files in os.walk(extract_path):
    level = root.replace(extract_path, '').count(os.sep)
    indent = ' ' * 2 * level
    print(f'{indent}{os.path.basename(root)}/')
    if level < 3:
        for f in files[:3]:
            print(f'{indent}  {f}')
        if len(files) > 3:
            print(f'{indent}  ... ({len(files)} dosya)')
```

```python
import shutil

src = '/content/idrid_extracted'
dst = '/content/drive/MyDrive/DR_Project/datasets/IDRiD'
os.makedirs(dst, exist_ok=True)

# Tum yapiyi kopyala (zaten kucuk dataset)
for item in os.listdir(src):
    s = os.path.join(src, item)
    d = os.path.join(dst, item)
    if os.path.isdir(s):
        shutil.copytree(s, d, dirs_exist_ok=True)
    else:
        shutil.copy2(s, d)

print("IDRiD Drive'a tasindi.")
print("Icerik:", os.listdir(dst))
```

```python
import os
from PIL import Image

idrid = '/content/drive/MyDrive/DR_Project/datasets/IDRiD'

# Segmentation training image sayisi
seg_train_img = None
for root, dirs, files in os.walk(idrid):
    if 'Original Images' in root and 'Training' in root:
        jpgs = [f for f in files if f.lower().endswith(('.jpg','.jpeg','.tif','.png'))]
        if jpgs:
            seg_train_img = root
            print(f"Training images path: {root}")
            print(f"Training image sayisi: {len(jpgs)}")

            # Ilk goruntu boyutu
            img = Image.open(os.path.join(root, jpgs[0]))
            print(f"Ornek boyut: {img.size}")
            break

# Lezyon maskelerini say
print("\nLezyon maske sayilari (training):")
for root, dirs, files in os.walk(idrid):
    if 'Groundtruths' in root and 'Training' in root:
        lesion = os.path.basename(root)
        masks = [f for f in files if f.lower().endswith(('.tif','.png','.jpg'))]
        if masks:
            print(f"  {lesion}: {len(masks)} maske")
```

```python
import os
import shutil

idrid_root = '/content/drive/MyDrive/DR_Project/datasets/IDRiD'

# 1) %20'leri bosluk yap
for item in os.listdir(idrid_root):
    if '%20' in item:
        old = os.path.join(idrid_root, item)
        new = os.path.join(idrid_root, item.replace('%20', ' '))
        os.rename(old, new)
        print(f"Yeniden adlandirildi: {item} -> {item.replace('%20', ' ')}")

# 2) Ic icine girmis cift klasor problemi (B. Disease Grading/B. Disease Grading/)
# Disardan tekrar ozdes klasor varsa duzelt
for top in os.listdir(idrid_root):
    top_path = os.path.join(idrid_root, top)
    if os.path.isdir(top_path):
        inner = os.path.join(top_path, top)
        if os.path.exists(inner):
            # Ic icine girmis - ic klasorun icerigini bir ust seviyeye tasi
            for child in os.listdir(inner):
                shutil.move(os.path.join(inner, child), os.path.join(top_path, child))
            os.rmdir(inner)
            print(f"Ic ice klasor cozuldu: {top}")

# 3) C. Localization kullanmayacagiz, sil (yer kazan)
loc_path = os.path.join(idrid_root, 'C. Localization')
if os.path.exists(loc_path):
    shutil.rmtree(loc_path)
    print("C. Localization silindi (gereksiz)")

print("\nFinal IDRiD yapisi:")
for top in sorted(os.listdir(idrid_root)):
    print(" -", top)
    sub = os.path.join(idrid_root, top)
    if os.path.isdir(sub):
        for s in sorted(os.listdir(sub))[:5]:
            print(f"     {s}")
```
