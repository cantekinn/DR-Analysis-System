```python
from pytorch_grad_cam import GradCAMPlusPlus
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget
from pytorch_grad_cam.utils.image import show_cam_on_image
print("Grad-CAM hazir.")
print(f"Gradio: {gr.__version__}")
```

```python
# Hucre 2: Inference + Gradio app
LESION_KEYS = ['MA', 'HE', 'EX', 'SE']
LESION_NAMES = ['Mikroanevrizma', 'Kanama', 'Sert Eksuda', 'Yumusak Eksuda']
LESION_COLORS = [(255, 0, 0), (255, 140, 0), (255, 255, 0), (0, 200, 255)]

stats = get_normalization_stats()
MEAN_T = torch.tensor(stats['mean']).view(1,3,1,1).to(device)
STD_T  = torch.tensor(stats['std']).view(1,3,1,1).to(device)

# Grad-CAM nesnesi
gradcam = GradCAMPlusPlus(model=clf, target_layers=[clf.conv_head])

def prep_classify(pil_img, size=384):
    """PIL -> RGB np uint8 (display) ve normalize tensor (model)."""
    arr = np.array(pil_img.convert('RGB'))
    # Disk yerine bellekte islemek icin: temp save degil, dogrudan fonksiyon
    from preprocessing import crop_fundus, apply_clahe, resize_square
    arr = crop_fundus(arr, tol=7)
    arr = apply_clahe(arr)
    arr = resize_square(arr, size=size)
    display = arr.astype(np.float32) / 255.0
    t = torch.from_numpy(arr).permute(2,0,1).unsqueeze(0).float().to(device) / 255.0
    t = (t - MEAN_T) / STD_T
    return t, display

def prep_segment(pil_img, size=512):
    arr = np.array(pil_img.convert('RGB'))
    from preprocessing import crop_fundus, apply_clahe, resize_square
    arr = crop_fundus(arr, tol=7)
    arr = apply_clahe(arr)
    arr = resize_square(arr, size=size)
    display = arr.astype(np.float32) / 255.0
    t = torch.from_numpy(arr).permute(2,0,1).unsqueeze(0).float().to(device) / 255.0
    t = (t - MEAN_T) / STD_T
    return t, display

def classify_image(pil_img):
    """Siniflandirma + Grad-CAM."""
    if pil_img is None:
        return None, "Lutfen bir gorunru yukleyin.", None
    t, display = prep_classify(pil_img)
    with torch.no_grad():
        logits = clf(t)
        probs = torch.softmax(logits, dim=1)[0].cpu().numpy()
    pred = int(np.argmax(probs))

    # Olasilik metni
    label_dict = {CLASS_NAMES[i]: float(probs[i]) for i in range(5)}

    # Grad-CAM (predicted class)
    targets = [ClassifierOutputTarget(pred)]
    grayscale_cam = gradcam(input_tensor=t, targets=targets)[0]
    cam_img = show_cam_on_image(display, grayscale_cam, use_rgb=True,
                                image_weight=0.55)

    # Tani metni
    severity_text = (
        f"### Sonuç: **{CLASS_NAMES[pred]}**\n\n"
        f"Model güveni: **{probs[pred]*100:.1f}%**\n\n"
    )

    return display, label_dict, cam_img

def segment_image(pil_img):
    """Segmentasyon + lezyon overlay."""
    if pil_img is None:
        return None, None, "Lutfen bir gorunru yukleyin."
    t, display = prep_segment(pil_img)
    with torch.no_grad():
        logits = seg(t)
        probs = torch.sigmoid(logits)[0].cpu().numpy()  # (4, H, W)
    pred = (probs > 0.5).astype(np.uint8)

    # Overlay
    overlay = display.copy()
    for i, color in enumerate(LESION_COLORS):
        m = pred[i] > 0
        c = np.array(color) / 255.0
        overlay[m] = c * 0.6 + overlay[m] * 0.4
    overlay = (overlay * 255).astype(np.uint8)

    # Per-lezyon piksel sayisi
    info = "### Tespit Edilen Lezyonlar\n\n"
    for i, key in enumerate(LESION_KEYS):
        n = int(pred[i].sum())
        info += f"- **{LESION_NAMES[i]} ({key}):** {n:,} piksel\n"

    return display, overlay, info

# ========== Gradio Interface ==========
with gr.Blocks(title="DR Analiz Sistemi", theme=gr.themes.Soft()) as demo:
    gr.Markdown("""
    # Diyabetik Retinopati Analiz Sistemi
    EEM0458 Görüntü İşleme Projesi · 2025-2026 Bahar · Can Tekin · BTÜ EEM

    Fundus (göz dibi) görüntüsü yükleyerek **5 sınıflı şiddet sınıflandırması**
    ve **4 lezyon tipi (Mikroanevrizma, Kanama, Sert/Yumuşak Eksuda) segmentasyonu**
    yapan derin öğrenme tabanlı sistemdir.
    """)

    with gr.Tabs():
        with gr.Tab("Şiddet Sınıflandırması (5 sınıf)"):
            with gr.Row():
                with gr.Column():
                    in1 = gr.Image(type="pil", label="Fundus Görüntüsü Yükle")
                    btn1 = gr.Button("Sınıflandır", variant="primary")
                with gr.Column():
                    out1_img = gr.Image(label="Ön İşlenmiş Görüntü")
                    out1_label = gr.Label(label="Sınıf Olasılıkları", num_top_classes=5)
                with gr.Column():
                    out1_cam = gr.Image(label="Grad-CAM++ Açıklanabilirlik")
            btn1.click(classify_image, inputs=in1,
                       outputs=[out1_img, out1_label, out1_cam])

        with gr.Tab("Lezyon Segmentasyonu (4 tip)"):
            with gr.Row():
                with gr.Column():
                    in2 = gr.Image(type="pil", label="Fundus Görüntüsü Yükle")
                    btn2 = gr.Button("Segmentasyon Yap", variant="primary")
                with gr.Column():
                    out2_img = gr.Image(label="Ön İşlenmiş Görüntü")
                    out2_overlay = gr.Image(label="Lezyon Overlay (MA-kırmızı, HE-turuncu, EX-sarı, SE-mavi)")
                with gr.Column():
                    out2_info = gr.Markdown(label="Tespit Bilgisi")
            btn2.click(segment_image, inputs=in2,
                       outputs=[out2_img, out2_overlay, out2_info])

    gr.Markdown("""
    ---
    **Not:** Bu sistem akademik bir prototiptir, klinik tanı için kullanılamaz.
    Modeller APTOS 2019 (sınıflandırma) ve IDRiD (segmentasyon) veri setleriyle eğitilmiştir.
    """)

# Public URL ile baslat (Colab icin share=True zorunlu)
demo.launch(share=True, debug=False)
```
